```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
der=pd.read_csv("cardata.csv")
der
```

```python
der.isnull().sum()
```

```python
der=der.dropna()
```

```python
der.isnull().sum()
```

```python
lpo=LabelEncoder()
der["size"]=lpo.fit_transform(der["size"])
der["total_sqft"]=lpo.fit_transform(der["total_sqft"])
```

```python
der
```

```python
der["size"]
```

```python
y=der[["price"]] #target value
y
```

```python
x=der[["size","total_sqft"]] # input features
x
```

```python
x.isnull().sum()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3 ,random_state=42)
```

```python
mode3=LinearRegression()
```
